Ejercicio 1: Lista de temperaturas diarias

```python
# Registrar temperaturas de una semana
temperaturas = [22.5, 24.0, 21.8, 23.2, 25.1, 20.5, 19.8]

# Calcular temperatura media
temperatura_media = sum(temperaturas) / len(temperaturas)
print(f"Temperatura media: {temperatura_media:.2f}°C")
```

Ejercicio 5: Lista de precios con IVA

```python
# Precios sin IVA
precios_sin_iva = [100, 50, 200, 75, 150]
IVA = 0.19  # 19%

# Calcular precios con IVA
precios_con_iva = [precio * (1 + IVA) for precio in precios_sin_iva]
print("Precios sin IVA:", precios_sin_iva)
print("Precios con IVA:", [round(precio, 2) for precio in precios_con_iva])
```

Ejercicio 6: Matriz identidad 3x3

```python
# Crear matriz identidad 3x3
matriz_identidad = []
for i in range(3):
    fila = []
    for j in range(3):
        if i == j:
            fila.append(1)
        else:
            fila.append(0)
    matriz_identidad.append(fila)

print("Matriz identidad 3x3:")
for fila in matriz_identidad:
    print(fila)
```

Ejercicio 7: Sumar dos listas

```python
# Dos listas del mismo tamaño
lista1 = [1, 2, 3, 4, 5]
lista2 = [10, 20, 30, 40, 50]

# Sumar elemento a elemento
suma_listas = [a + b for a, b in zip(lista1, lista2)]
print("Lista 1:", lista1)
print("Lista 2:", lista2)
print("Suma:", suma_listas)
```

Ejercicio 10: Transponer una matriz

```python
# Matriz original 3x3
matriz_original = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Transponer la matriz
matriz_transpuesta = [[fila[i] for fila in matriz_original] for i in range(len(matriz_original[0]))]

print("Matriz original:")
for fila in matriz_original:
    print(fila)

print("Matriz transpuesta:")
for fila in matriz_transpuesta:
    print(fila)
```

Ejercicio 11: Eliminar duplicados

```python
# Lista con elementos repetidos
lista_con_duplicados = [1, 2, 2, 3, 4, 4, 4, 5, 1, 6]

# Eliminar duplicados
lista_sin_duplicados = list(set(lista_con_duplicados))

print("Lista con duplicados:", lista_con_duplicados)
print("Lista sin duplicados:", lista_sin_duplicados)
```

Ejercicio 12: Contador de vocales

```python
# Frase para analizar
frase = "Python es un lenguaje de programación muy poderoso"

# Contar vocales
vocales = "aeiouAEIOU"
contador = 0

for caracter in frase:
    if caracter in vocales:
        contador += 1

print(f"Frase: '{frase}'")
print(f"Número de vocales: {contador}")
```

Ejercicio 13: Matriz de multiplicación

```python
# Crear tabla de multiplicar 5x5
tabla_multiplicar = []
for i in range(1, 6):
    fila = []
    for j in range(1, 6):
        fila.append(i * j)
    tabla_multiplicar.append(fila)

print("Tabla de multiplicar 5x5:")
for fila in tabla_multiplicar:
    print(fila)
```

Ejercicio 14: Reemplazar elementos negativos

```python
# Lista con números positivos y negativos
numeros = [5, -2, 10, -8, 3, -1, 7]

# Reemplazar negativos por cero
numeros_sin_negativos = [num if num >= 0 else 0 for num in numeros]

print("Lista original:", numeros)
print("Lista sin negativos:", numeros_sin_negativos)
```

Ejercicio 15: Crear matriz de caracteres

```python
# Crear matriz 3x3 con letras del alfabeto
letra_actual = 65  # Código ASCII de 'A'
matriz_letras = []

for i in range(3):
    fila = []
    for j in range(3):
        fila.append(chr(letra_actual))
        letra_actual += 1
    matriz_letras.append(fila)

print("Matriz de letras 3x3:")
for fila in matriz_letras:
    print(fila)
```
